```python
import glob
import json
import matplotlib.pyplot as plt
import numpy as np

# %matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'



val_acc = []
train_acc = []

epocs = []
hidden = []
learn = []
reg = []


for file in glob.glob("*.json"):
    with open(file, 'r') as infile:
        run_details = json.load(infile)
        epocs.append(run_details['epocs'])
        hidden.append(run_details['hidden'])
        learn.append(run_details['learning_rate'])
        reg.append(run_details['reg'])
        train_acc.append(run_details['train_acc_history'][-1])
        val_acc.append(run_details['val_acc_history'][-1])
#         print file, run_details['val_acc_history'][-1]
  

print np.max(val_acc)

# plt.subplot(4, 1, 1)
plt.scatter(epocs, val_acc, c='b')
plt.xlabel('epocs')
plt.ylabel('accuracy')
plt.title("epocs")
plt.show()

# plt.subplot(4, 1, 2)
plt.scatter(hidden, val_acc, c='b')
plt.xlabel('hidden')
plt.ylabel('accuracy')
plt.title("hidden")
plt.show()


# plt.subplot(4, 1, 3)
plt.scatter(np.log10(learn), val_acc, c='b')
plt.xlabel('learn')
plt.ylabel('accuracy')
plt.title("learning rate")
plt.show()


# plt.subplot(4, 1, 4)
plt.scatter(np.log10(reg), val_acc, c='b')
plt.xlabel('reg')
plt.ylabel('accuracy')
plt.title("regularization")
plt.show()
```
