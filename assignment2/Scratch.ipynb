```python
def forward(x, beta, gamma, eps):

    mean = np.mean(x)    
    x_minus_mean = x - mean
    squares = x_minus_mean ** 2
    var = np.mean(squares, axis=0)
    sqrt_var = np.sqrt(var + eps)
    one_over_sqrt_var = 1 / sqrt_var
    xhat = one_over_sqrt_var * x_minus_mean
    return gamma * xhat + beta

N, D = 3, 2
x = np.array(range(1, N*D+1), dtype="float64").reshape(N, D)
gamma = 3 * np.ones(D)
beta =  2 * np.ones(D)
eps = 1e-7

dout = np.ones_like(x)

print "x\n", x
print "beta\n", beta
print "gamma\n", gamma
print "forward\n", forward(x, beta, gamma, eps)

# evaluate numerical gradient
fx = lambda x: forward(x, beta, gamma, eps)
fbeta = lambda beta: forward(x, beta, gamma, eps)
fgamma = lambda gamma: forward(x, beta, gamma, eps)

h = 1e-3
dx_num = eval_numerical_gradient_array(fx, x, dout, h)
dbeta_num = eval_numerical_gradient_array(fbeta, beta, dout, h)
dgamma_num = eval_numerical_gradient_array(fgamma, gamma, dout, h)
print "dx_num\n", dx_num
print "dbeta_num\n", dbeta_num
print "dgamma_num\n", dgamma_num

# Evaluate gradient with forward and backprop and compare to
# numerical estimate

# forward pass as needed
mean = np.mean(x)    
x_minus_mean = x - mean
squares = x_minus_mean ** 2
var = np.mean(squares, axis=0)
sqrt_var = np.sqrt(var + eps)
one_over_sqrt_var = 1 / sqrt_var
xhat = one_over_sqrt_var * x_minus_mean
y = gamma * xhat + beta

# 9 + beta
dbeta = np.sum(dout, axis=0)
dgamma_xhat = dout

# 8 xhat * gamma
dgamma = np.sum(dgamma_xhat * xhat, axis=0)
dxhat = dgamma_xhat * gamma

# 7 x_minus_mean * one_over_sqrt_var
done_over_sqrt_var = np.sum(dxhat * x_minus_mean, axis=0)
dx_minus_mean_top = dxhat * one_over_sqrt_var

# 6 1/x
dsqrt_var = -(sqrt_var ** (-2)) * done_over_sqrt_var

#5 sqrt(var + eps)
dvar = 0.5 * ((var + eps)**(-0.5)) * dsqrt_var 

#4 np.mean
dsquares = (1.0 / N) * np.ones((N, D)) * dvar 

# 3 x**2
dx_minus_mean_bottom = 2.0 * x_minus_mean * dsquares

# 2 (x - mean)
# sum inbound gradient
dx_minus_mean = dx_minus_mean_top + dx_minus_mean_bottom 
# gradient to input node
dx_input = dx_minus_mean
# gradient to mean node
dmean = -np.sum(dx_minus_mean, axis=0)

#1 mean(x)
dx_mean = (1.0 / N) * np.ones((N, D)) * dmean

dx = dx_input + dx_mean

# compare
print "rel error dbeta", rel_error(dbeta, dbeta_num)
print "rel error dgamma", rel_error(dgamma, dgamma_num)
print "rel error dx", rel_error(dx, dx_num)
print dx_num
print dx
```
